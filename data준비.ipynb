```python
from gensim.models import Word2Vec
import re
```

```python
sentences = ["Homer simpson forgot his lunch at home, so he had to buy a burger on his way to work"]
```

```python
sentences = [re.sub('[^a-zA-Z]', ' ', sentence).lower().split(" ") for sentence in sentences]
```

```python
sentences[0]
```

```python
skip_gram = Word2Vec(sentences, vector_size=40,min_count=1, window=5, sg=1)
```

```python
print(skip_gram.wv.get_vector(skip_gram.wv.key_to_index['homer']))
```
